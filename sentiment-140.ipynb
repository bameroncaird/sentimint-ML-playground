```python
# manually added column labels in csv file
csv = pd.read_csv(file_path, encoding='latin1')
csv
```

```python
# check distribution of labels
labels = csv[['target']]

num0 = 0; num2 = 0; num4 = 0; other = 0

for i in range(len(labels)):
    label = labels.loc[i][0]
    
    if label == 0:
        num0 += 1
    elif label == 2:
        num2 += 1
    elif label == 4:
        num4 += 1
    else:
        other += 1
```

```python
# there are no neutral labels
print(num0, num2, num4, other)
```
